```python
# Numerical arrays.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt
```

## Simple Plot

```python
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
np.linspace(0.0, 10.0, 101)
```

```python
# https://numpy.org/doc/stable/reference/generated/numpy.arange.html
np.arange(0.0, 10.1, 0.1)
```

```python
# X values.
x = np.linspace(0.0, 10.0, 101)
```

$y = mx + c$

```python
# Y values.
y = 3.0 * x + 1.0

# Have a look.
y
```

```python
# Plot x vs y.
plt.plot(x, y, label='y = 3x + 1')

# Add a title.
plt.title("A Basic Line")

# X axis label.
plt.xlabel('x')

# Y axis label.
plt.ylabel('y')

# Add a legend.
plt.legend();
```

## More Than One Plot

```python
# X values.
x = np.linspace(0.0, 100.0, 101)

# Plot.
plt.plot(x, x, label='x')
plt.plot(x, x**2, label='x^2')
plt.plot(x, 2**x, label='2^x')

# Add a title.
plt.title("Comparing Functions")

# X axis label.
plt.xlabel('x')

# Add a legend.
plt.legend();
```
