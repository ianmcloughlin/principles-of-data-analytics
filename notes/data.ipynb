## Reading CSV Files in NumPy

https://numpy.org/doc/stable/user/how-to-io.html

https://numpy.org/doc/stable/reference/generated/numpy.loadtxt.html#numpy-loadtxt

```python
# Read data from a CSV using numpy.
data = np.loadtxt("../data/aapl.csv", skiprows=1, delimiter=",", usecols=(1, 2, 3, 4, 5, 6))

# Have a look.
data
```

## Slicing

```python
# The first row.
data[0]
```

```python
# The first column.
data[:, 0]
```
