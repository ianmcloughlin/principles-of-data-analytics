```python
# Open a file.
with open('binary-and-text.ipynb', 'rb') as f:
  b = f.read(1)
```

```python
# See the byte in decimal.
int(b.hex(), 16)
```

```python
# See the byte in binary.
f'{int(b.hex(), 16):08b}'
```
