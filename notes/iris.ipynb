# Iris Data Set

```python
# Data frames.
import pandas as pd

# Plotting.
import matplotlib.pyplot as plt
```

```python
# Load the iris data set.
df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/refs/heads/main/pandas/tests/io/data/csv/iris.csv")
```

```python
# Have a look.
df
```

## Histogram

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
# Create a histogram.
plt.hist(df['SepalLength'])
```

## Box Plot

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
# Create a box plot.
plt.boxplot(df['SepalWidth'])
```

# Scatter Plot

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
# Plot.
plt.plot(df['PetalLength'], df['PetalWidth'], 'k.')
```
